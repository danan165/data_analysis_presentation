## Python vs. R

### Python for Data Analysis

* General-purpose language
* Emphasis on code readability and productivity
* Relies on packages

### When to use Python?

* Repeated/Automated tasks
* Integrate data analysis tasks into...
    * web apps
    * production databases
* Implement algorithms for production use
* Integrate with every part of workflow

### R for Data Analysis

* Developed for statisticians
* Supports basic data analysis without packages
* Designed for data analysis
* Convenient for data analysis:
    * Vast amount of packages
    * Readily usable tests
    * Use of formulas

### When to use R?

* Exploratory data analysis
* Answer specific business question
* Task requires standalone computing
* Analysis on individual servers

## Utilizing the Data Analysis Process: R Demo

**We are looking to list an Airbnb room in Boston. Based on the dataset, what kind of room should we list?**

### Useful Terms/Definitions

* **Observations**: the rows in a table or dataframe. 
* **Variables**: the columns in a table.
    * **Qualitative Variables**: no natural order, but an order can be imposed.
    * **Quantitative Variables**: numeric values that have a natural sorting order.

### Types of Variable Analysis

**Methods to guide us throughout the data analysis process.**

### Qualitative Univariate Analysis

* **Univariate Analysis**: analysis of a single variable.
* **Qualitative Univariate Analysis**: analysis of a single qualitative (categorical) variable.

**What is the most popular room type in Boston?**

```python
%R data.frame(table(boston$room_type))
```

**Which neighborhood in Boston has the most rooms listed?**

```python
%R data.frame(table(boston$neighborhood))
```

```python
%%R 

barplot(
    table(boston$neighborhood), 
    ylim = c(0, 500), 
    las = 2, 
    ylab = 'Room Quantity', 
    main='Room Quantity by Neighborhood'
)
```

### Quantitative and Qualitative Bivariate Analysis

* **Bivariate Analysis**: analysis of two variables to determine a relationship between them.
* **Qualitative and Quantitative Bivariate Analysis**: relationship between a numeric variable's values across multiple categories.

**Which neighborhood in Boston has the most reviews submitted?**

```python
%R data.frame('Reviews per Neighborhood'=tapply(boston$reviews, boston$neighborhood, FUN = sum))
```

```python
%%R 
barplot(
  tapply(boston$reviews, boston$neighborhood, FUN = sum),
  las = 2,
  ylim = c(0, 14000),
  main = 'Review Quantity by Neighborhood'
)
```

### Sort and Subset Operations

**What are the most active neighborhoods, based on reviews submitted and rooms listed?**

```python
%%R

neighborhoods_by_rooms <- 
head(
  sort(
    table(boston$neighborhood), 
    decreasing = TRUE
  ), 
  n = 5
)
```

```python
%%R

neighborhoods_by_reviews <- 
head(
  sort(
    tapply(boston$reviews, boston$neighborhood, FUN = sum), 
    decreasing = TRUE
  ), 
  n =5
)
```

```python
%%R

active_neighborhoods <- 
  intersect(names(neighborhoods_by_reviews), names(neighborhoods_by_rooms))
```

```python
%R active_neighborhoods
```

### Quantitative and Qualitative Bivariate Analysis

* **Bivariate Analysis**: analysis of two variables to determine a relationship between them.
* **Qualitative and Quantitative Bivariate Analysis**: relationship between a numeric variable's values across multiple categories.

**What are the review results from these active neighborhoods?**

```python
%%R

summaries <- data.frame(
    'Jamaica Plain' = unclass(summary(boston$overall_satisfaction[boston$reviews > 0 & boston$neighborhood=='Jamaica Plain'])), 
    'South End'=unclass(summary(boston$overall_satisfaction[boston$reviews > 0 & boston$neighborhood=='South End'])), 
    'Dorchester'=unclass(summary(boston$overall_satisfaction[boston$reviews > 0 & boston$neighborhood=='Dorchester'])), 
    'Back Bay'=unclass(summary(boston$overall_satisfaction[boston$reviews > 0 & boston$neighborhood=='Back Bay']))
)
```

```python
%R summaries
```

```python
%%R

par(mfrow = c(2, 2))
for(nb in active_neighborhoods) {
  hist(
    x = boston$overall_satisfaction[boston$reviews > 0 
                                    & boston$neighborhood==nb],
    ylim = c(0, 200),
    xlab = "Review Results",
    main = paste(nb, " Review Results")
  )
}
```

**What is the price distribution in Jamaica Plain and South End for an Entire home/apt?**

```python
%%R 
prices <- data.frame(
    'Jamaica Plain'=unclass(summary(boston$price[boston$neighborhood=='Jamaica Plain' & boston$room_type=='Entire home/apt'])),
    'South End'=unclass(summary(boston$price[boston$neighborhood=='South End' & boston$room_type=='Entire home/apt']))
)
```

```python
%R prices
```

```python
%%R

par(mfrow = c(1, 2))
hist(
  x=boston$price[boston$room_type=="Entire home/apt" &
                   boston$neighborhood=="South End"],
  xlab = "price",
  breaks = 20,
  main = "South End Entire home/apt",
  ylim = c(0, 100)
)

hist(
  x=boston$price[boston$room_type=="Entire home/apt" & 
                   boston$neighborhood=="Jamaica Plain"],
  xlab = "price",
  breaks = 20,
  main = "Jamaica Plain Entire home/apt",
  ylim = c(0, 60)
)
```

### Demo Conclusions

* Room Type: Entire home/apt
* Neighborhood: Jamaica Plain or South End
* Price: $120-250
